```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup=requests.get(url)
soup
```

```python
lxml_code=BeautifulSoup(soup.text, "lxml")
soup.text
lxml_code
```

```python
table_code = lxml_code.table
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
data= []
for tag in tags:
  data.append(tag.text.split('\n')[1:-1])
data  
```

```python
import csv
f=open('list_data.csv','w')
x=csv.writer(f)
for i in data:
  x.writerow(i)

f.close()
```
